# Import Libraries

```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
#import codecademylib3

# load in the data
df = pd.read_csv("/Users/user/Projects_CodeAcademy/Mushrooms/mushrooms.csv")
print(df.head())

# list of all column headers
columns = df.columns.tolist()
```

Qou will see that every single variable is a categorical variable. Therefore, you can create bar charts for the counts of every variable in the csv file.

## Bar charts

```python
# For loops to iterate over all columns

for column in columns:
    # print('Column Name : ', columns)
    sns.countplot(df[column], order=df[column].value_counts().index)
    # rotates the value labels slightly so they dont overlap, also slightly increases font size
    plt.xticks(rotation=30, fontsize=10)
# increases the variable label font size slightly to increase readability
    plt.xlabel(column, fontsize=12)
    plt.title(column + " - Value counts")

    plt.show()
    plt.clf()
```
