## How does the distribution of accounts look like across different countries?

Let us first create the two subsets for low and middle-and high income countries

```python
## High income
data_account2 = new_data[new_data['Indicator Name'].isin(['Account, female (% age 15+)', 'Account, male (% age 15+)'])]

high_data = data_account2[data_account2['Country Name'].isin(['High income'])]
#print(high_data)

# Drop columns that we do not need
high2 = high_data.drop(columns=['Unnamed: 7','Country Name', 'Indicator Code', 'Country Code', 'Indicator Code'],  axis=1)


# Change the value names for indicator to reflect high income countries
high2["Indicator Name"].replace({"Account, female (% age 15+)": "Account(F): High income", "Account, male (% age 15+)": "Account(M): High income"}, inplace=True)
#print(high2)

#high = high2.set_index('Indicator Name').T.plot()


## Low income
data_account2 = new_data[new_data['Indicator Name'].isin(['Account, female (% age 15+)', 'Account, male (% age 15+)'])]

low_data = data_account2[data_account2['Country Name'].isin(['Low & middle income'])]
#print(low_data)

# Drop columns that we do not need
low2 = low_data.drop(columns=['Unnamed: 7','Country Name', 'Indicator Code', 'Country Code', 'Indicator Code'],  axis=1)
#print(low2)


low2["Indicator Name"].replace({"Account, female (% age 15+)": "Account(F): Low income", "Account, male (% age 15+)": "Account(M): Low income"}, inplace=True)
#print(low2)



# Combine the dataframes

frames = [low2, high2]
result = pd.concat(frames)
print(result)
```

Now, it's time to plot the datasets in one single graph

```python
result.set_index('Indicator Name').T.plot()
```
